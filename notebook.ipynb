```python
import pandas as pd

# Load the datasets
train_df = pd.read_csv('train_dataset.csv')
test_df = pd.read_csv('test_dataset.csv')

# Display the first few rows of the training dataset to understand its structure
train_df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import pickle

#Conjuntos de entrenamiento y prueba
X_train = train_df.drop(columns='target')
X_test = test_df.drop(columns='target')
y_train = train_df['target']
y_test = test_df['target']

# Entrenar un modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

#Evaluar el modelo en el conjunto de prueba
y_pred = model.predict(X_test)

# Guardar el modelo en un archivo 'model.pkl'
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)

```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Evaluar el modelo en el conjunto de datos de prueba
y_train_pred = model.predict(X_train)

# Calcular métricas para el conjunto de entrenamiento
mse_train = mean_squared_error(y_train, y_train_pred)
mae_train = mean_absolute_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

# Calcular métricas para el conjunto de prueba
mse_test = mean_squared_error(y_test, y_pred)
mae_test = mean_absolute_error(y_test, y_pred)
r2_test = r2_score(y_test, y_pred)

print(f"""Métricas conjunto de entrenamiento:\nMSE (Error Cuadrático Medio): {mse_train}
MAE (Error Absoluto Medio): {mae_train}
R² (Coeficiente de Determinación): {r2_train}

Métricas conjunto de prueba:
MSE (Error Cuadrático Medio): {mse_test}
MAE (Error Absoluto Medio): {mae_test}
R² (Coeficiente de Determinación): {r2_test}""")
```
